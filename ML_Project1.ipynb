```python
import pandas as pd
import numpy as np

# تعداد داده‌ها
data_count = 500 # افزایش تعداد برای دقت بیشتر مدل

data = {
    'Age': np.random.randint(8, 16, data_count),
    'Height': np.random.normal(160, 12, data_count),
    'Arm_Span': np.random.normal(162, 13, data_count),
    'Agility_Test': np.random.normal(17, 2.5, data_count),
    'Reaction_Time': np.random.uniform(0.15, 0.45, data_count)
}

df = pd.DataFrame(data)

# تعریف منطق استعدادیابی (بر اساس همبستگی قد و چابکی)
# در تنیس، ترکیب قد بلند و چابکی بالا امتیاز مثبت است
df['Tennis_Talent'] = ((df['Height'] > 155) & (df['Agility_Test'] < 18)).astype(int)

df.to_csv('iranian_tennis_talent.csv', index=False)
print("دیتاست معتبر با 500 نمونه ساخته شد.")
```

```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
import joblib

# ۱. ساخت دیتاست هوشمندتر
data_count = 1000
np.random.seed(42)

ages = np.random.randint(8, 17, data_count)
heights = np.random.normal(155, 15, data_count)
agilities = np.random.normal(16, 3, data_count)
reactions = np.random.uniform(0.1, 0.5, data_count)

# منطق استعدادیابی: برای سنین پایین‌تر، سخت‌گیری در چابکی کمتر است
# یعنی کودک 8 ساله با چابکی 15 ثانیه مستعد است، اما نوجوان 16 ساله با 15 ثانیه معمولی است
talent_score = (heights > 150) & (agilities < (12 + (ages-8)*0.5)) & (reactions < 0.3)
labels = talent_score.astype(int)

df = pd.DataFrame({
    'Age': ages, 'Height': heights,
    'Agility': agilities, 'Reaction': reactions,
    'Target': labels
})

model = DecisionTreeClassifier(max_depth=4)
model.fit(df.drop('Target', axis=1), df['Target'])


joblib.dump(model, 'tennis_expert_model.pkl')
print("مدل هوشمند متنی با موفقیت ساخته و ذخیره شد.")
```

```python
import numpy as np
from sklearn.ensemble import RandomForestClassifier
import joblib

# [gender, age, height, arm, agility, reaction]
X = np.array([
    [1, 15, 170, 180, 11, 0.22],
    [0, 14, 160, 165, 13, 0.30],
    [1, 12, 150, 155, 16, 0.40],
    [0, 13, 155, 160, 14, 0.35],
    [1, 16, 175, 185, 10, 0.20]
])

y = np.array([1, 1, 0, 0, 1])  # 1 = استعداد تنیس

model = RandomForestClassifier()
model.fit(X, y)

joblib.dump(model, "tennis_model_v1.pkl")
print("OK")
```

```python
import gradio as gr
import joblib

model = joblib.load('tennis_model_v1.pkl')

def predict(gender, age, height, arm, agility, reaction):
    # تبدیل جنسیت به عدد
    gender_val = 1 if gender == "مرد" else 0

    pred = model.predict([[gender_val, age, height, arm, agility, reaction]])

    if pred[0] == 1:
        return "✅ واجد شرایط برای آکادمی تخصصی تنیس"
    else:
        return "⚠️ پیشنهاد می‌شود در سایر رشته‌های ورزشی تست دهید"


interface = gr.Interface(
    fn=predict,
    inputs=[
        gr.Radio(["مرد", "زن"], label="جنسیت"),
        gr.Slider(8, 18, label="سن (سال)"),
        gr.Number(label="قد (سانتی‌متر)"),
        gr.Number(label="طول دو دست (سانتی‌متر)"),
        gr.Slider(10, 25, label="زمان تست چابکی (ثانیه)"),
        gr.Slider(0.1, 0.6, label="زمان عکس‌العمل (ثانیه)")
    ],
    outputs="text",
    title="سامانه استعدادیابی ورزشی ایران",
    theme="soft"
)

interface.launch()
```

```python
import pandas as pd
import numpy as np

# تعداد داده‌ها
data_count = 500 # افزایش تعداد برای دقت بیشتر مدل

data = {
    'Age': np.random.randint(8, 16, data_count),
    'Height': np.random.normal(160, 12, data_count),
    'Arm_Span': np.random.normal(162, 13, data_count),
    'Agility_Test': np.random.normal(17, 2.5, data_count),
    'Reaction_Time': np.random.uniform(0.15, 0.45, data_count)
}

df = pd.DataFrame(data)

# تعریف منطق استعدادیابی (بر اساس همبستگی قد و چابکی)
# در تنیس، ترکیب قد بلند و چابکی بالا امتیاز مثبت است
df['Tennis_Talent'] = ((df['Height'] > 155) & (df['Agility_Test'] < 18)).astype(int)

df.to_csv('iranian_tennis_talent.csv', index=False)
print("دیتاست معتبر با 500 نمونه ساخته شد.")
```

```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
import joblib

# ۱. ساخت دیتاست هوشمندتر
data_count = 1000
np.random.seed(42)

ages = np.random.randint(8, 17, data_count)
heights = np.random.normal(155, 15, data_count)
agilities = np.random.normal(16, 3, data_count)
reactions = np.random.uniform(0.1, 0.5, data_count)

# منطق استعدادیابی: برای سنین پایین‌تر، سخت‌گیری در چابکی کمتر است
# یعنی کودک 8 ساله با چابکی 15 ثانیه مستعد است، اما نوجوان 16 ساله با 15 ثانیه معمولی است
talent_score = (heights > 150) & (agilities < (12 + (ages-8)*0.5)) & (reactions < 0.3)
labels = talent_score.astype(int)

df = pd.DataFrame({
    'Age': ages, 'Height': heights,
    'Agility': agilities, 'Reaction': reactions,
    'Target': labels
})

# ۲. آموزش مدل سبک (برای نمره حجم کم)
model = DecisionTreeClassifier(max_depth=4)
model.fit(df.drop('Target', axis=1), df['Target'])

# ۳. ذخیره مدل
joblib.dump(model, 'tennis_expert_model.pkl')
print("مدل هوشمند متنی با موفقیت ساخته و ذخیره شد.")
```

```python
import numpy as np
from sklearn.ensemble import RandomForestClassifier
import joblib

# دیتای ساختگی اما منطقی (برای پروژه)
X = np.array([
    [12, 150, 155, 14, 0.30],
    [15, 170, 180, 11, 0.22],
    [10, 140, 140, 18, 0.45],
    [14, 165, 170, 12, 0.25],
    [13, 155, 160, 15, 0.35]
])

y = np.array([1, 1, 0, 1, 0])  # 1 = استعداد تنیس

model = RandomForestClassifier()
model.fit(X, y)

joblib.dump(model, 'tennis_model_v1.pkl')
print("✅ مدل ساخته و ذخیره شد")
```
